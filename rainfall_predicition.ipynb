### Importing the Dataset
We import the "Weather_Data.csv" dataset from a specified URL using the *requests* library. It retrieves the data from the URL and saves it as a local file named "Weather_Data.csv". Then we read the data into a dataframe using pandas.

```python
import requests

def download(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, "wb") as f:
            f.write(response.content)

path = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillUp/labs/ML-FinalAssignment/Weather_Data.csv'
filename = "Weather_Data.csv"
download(path, filename)
```

```python
df = pd.read_csv("Weather_Data.csv")
df.head()
```

```python
df.WindGustSpeed.dtype
```

### Data Preprocessing


#### One Hot Encoding


First, we need to perform one hot encoding to convert categorical variables to binary variables.


```python
df_sydney_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

Next, we replace the values of the 'RainTomorrow' column changing them from a categorical column to a binary column. We do not use the `get_dummies` method because we would end up with two columns for 'RainTomorrow' and we do not want, since 'RainTomorrow' is our target.


```python
df_sydney_processed.replace(['No', 'Yes'], [0,1], inplace=True)
```

### Training Data and Test Data


Now, we set our 'features' or x values and our Y or target variable.


```python
df_sydney_processed.drop('Date',axis=1,inplace=True)
```

```python
df_sydney_processed = df_sydney_processed.astype(float)
```

```python
features = df_sydney_processed.drop(columns='RainTomorrow', axis=1)
Y = df_sydney_processed['RainTomorrow']
```

### Linear Regression


```python
x_train, x_test, y_train, y_test = train_test_split(features, Y, test_size=0.2, random_state=10)
```

**Creating and training a Linear Regression model called LinearReg using the training data (`x_train`, `y_train`).**


```python
# Create a Linear Regression model
LinearReg = LinearRegression()

# Train the model
LinearReg.fit(x_train, y_train)
```

 **Using the `predict` method on the testing data (`x_test`) and saving it to the array `predictions`.**


```python
predictions_lr = LinearReg.predict(x_test)
```

**Using the `predictions` and the `y_test` dataframe to calculate the value for each metric using the appropriate function.**


```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calculate mean squared error (MSE)
LinearRegression_MSE = mean_squared_error(y_test, predictions_lr)

# Calculate mean absolute error (MAE)
LinearRegression_MAE = mean_absolute_error(y_test, predictions_lr)

# Calculate coefficient of determination (R-squared)
LInearRegression_R2 = r2_score(y_test, predictions_lr)
```

**Showing the MAE, MSE, and R2 in a tabular format using data frame for the linear model.**


```python
# Create a DataFrame to store the metrics
LinReg_report = pd.DataFrame({'Metric': ['LinearRegression_MAE', 'LinearRegression_MSE', 'LInearRegression_R2'],
                           'Score': [LinearRegression_MAE, LinearRegression_MSE, LInearRegression_R2]})

# Display the DataFrame
LinReg_report.head()
```

###  K-Nearest Neighbors


**Creating and training a KNN model called KNN using the training data (`x_train`, `y_train`) with the `n_neighbors` parameter set to `4`.**


```python
from sklearn.neighbors import KNeighborsRegressor

# Create a KNN model
KNN = KNeighborsRegressor(n_neighbors=4)

# Train the model
KNN.fit(x_train, y_train)
```

**Using the predict method on the testing data (x_test) and saving it to the array predictions.**

```python
predictions_knn = KNN.predict(x_test)
```

**Using the predictions and the y_test dataframe to calculate the value for each metric using the appropriate function.**

```python
from sklearn.metrics import accuracy_score, jaccard_score, f1_score

# Convert the predictions to discrete values for classification
predictions_discrete = (predictions_knn > 0.5).astype(int)

# Calculate the accuracy score
KNN_Accuracy_Score = accuracy_score(y_test, predictions_discrete)

# Calculate the Jaccard index
KNN_JaccardIndex = jaccard_score(y_test, predictions_discrete)

# Calculate the F1 score
KNN_F1_Score = f1_score(y_test, predictions_discrete)
```

**Showing the Accuracy_Score, JaccardIndex, and F1_Score in a tabular format using data frame for the K-Nearest Neighbors model.**

```python
# Create a DataFrame to store the metrics
KNN_Report = pd.DataFrame({'Metric': ['KNN_Accuracy_Score', 'KNN_JaccardIndex', 'KNN_F1_Score'],
                           'Score': [KNN_Accuracy_Score, KNN_JaccardIndex, KNN_F1_Score]})

# Display the DataFrame
KNN_Report.head()
```

### Decision Tree


**Creating and training a Decision Tree model called Tree using the training data (`x_train`, `y_train`).**


```python
from sklearn.tree import DecisionTreeRegressor

# Create a Decision Tree model
Tree = DecisionTreeRegressor()

# Train the model
Tree.fit(x_train, y_train)
```

**Using the predict method on the testing data (x_test) and saving it to the array predictions.**

```python
# Using the trained model to make predictions on the testing data
predictions_dt = Tree.predict(x_test)
```

**Using the predictions and the y_test dataframe to calculate the value for each metric using the appropriate function.**

```python

# Convert predictions to discrete class labels
predictions_discrete = (predictions_dt > 0.5).astype(int)

# Calculate the accuracy score
Tree_Accuracy_Score = accuracy_score(y_test, predictions_discrete)

# Calculate the Jaccard index
Tree_JaccardIndex = jaccard_score(y_test, predictions_discrete)

# Calculate the F1 score
Tree_F1_Score = f1_score(y_test, predictions_discrete)
```

**Showing the Accuracy_Score, JaccardIndex, and F1_Score in a tabular format using data frame for the Decision Tree model.**

```python
# Create a DataFrame to store the metrics
Tree_Report = pd.DataFrame({'Metric': ['Tree_Accuracy_Score', 'Tree_JaccardIndex', 'Tree_F1_Score'],
                           'Score': [Tree_Accuracy_Score, Tree_JaccardIndex, Tree_F1_Score]})

# Display the DataFrame
Tree_Report.head()
```

### Logistic Regression


**Using the `train_test_split` function to split the `features` and `Y` dataframes with a `test_size` of `0.2` and the `random_state` set to `1`.**


```python
from sklearn.model_selection import train_test_split

# Assuming you have your feature dataframe stored in the variable 'features'
# Assuming you have your target dataframe stored in the variable 'Y'

# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features, Y, test_size=0.2, random_state=1)
```

**Creating and training a LogisticRegression model called LR using the training data (`x_train`, `y_train`) with the `solver` parameter set to `liblinear`.**


```python
from sklearn.linear_model import LogisticRegression

# Create a Logistic Regression model
LR = LogisticRegression(solver='liblinear')

# Train the model
LR.fit(X_train, y_train)
```

**Using the `predict` and `predict_proba` methods on the testing data (`x_test`) and saving it as 2 arrays `predictions` and `predict_proba`.**


```python
# Use the trained model to make predictions on the testing data
predictions = LR.predict(X_test)

# Use the trained model to predict class probabilities on the testing data
predict_proba = LR.predict_proba(X_test)
```

**Using the `predictions`, `predict_proba` and the `y_test` dataframe to calculate the value for each metric using the appropriate function.**


```python

from sklearn.metrics import accuracy_score, jaccard_score, f1_score

# Calculate accuracy score
LR_Accuracy_Score = accuracy_score(y_test, predictions)

# Calculate Jaccard index
LR_JaccardIndex = jaccard_score(y_test, predictions)

# Calculate F1 score
LR_F1_Score = f1_score(y_test, predictions)

# Calculate Log loss
LR_Log_Loss = log_loss(y_test, predict_proba)
```

**Showing the Accuracy_Score, JaccardIndex, F1_Score, and Log_Loss in a tabular format using data frame for the logistic model.**

```python
# Create a DataFrame to store the metrics
LogReg_Report = pd.DataFrame({'Metric': ['LR_Accuracy_Score', 'LR_JaccardIndex', 'LR_F1_Score', 'LR_Log_Loss' ],
                           'Score': [LR_Accuracy_Score, LR_JaccardIndex, LR_F1_Score, LR_Log_Loss]})

# Display the DataFrame
LogReg_Report.head()
```

**Showing the Accuracy_Score, JaccardIndex, and F1_Score in a tabular format using data frame for the SVM model.**
